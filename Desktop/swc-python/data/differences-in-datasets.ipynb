```python
import glob
import numpy
import matplotlib.pyplot

filenames = sorted(glob.glob('inflammation*.csv'))

data0 = numpy.loadtxt(fname=filenames[0], delimiter=',')
data1 = numpy.loadtxt(fname=filenames[1], delimiter=',')

fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0))

matplotlib.pyplot.ylabel('Difference in average')
matplotlib.pyplot.plot(numpy.mean(data0, axis=0) - numpy.mean(data1, axis=0))

fig.tight_layout()
matplotlib.pyplot.show()
```

```python
import glob
import numpy as np
import matplotlib.pyplot as plt

filenames = sorted(glob.glob('inflammation*.csv'))
plt.figure(figsize=(10.0, 3.0))
plt.ylabel('Difference in average')

for i in range(len(filenames)-1):
    data0 = np.loadtxt(fname=filenames[i], delimiter=',')
    data1 = np.loadtxt(fname=filenames[i+1], delimiter=',')
    plt.plot(np.mean(data0, axis=0) - np.mean(data1, axis=0), label=f'{i} and {i+1}')
    plt.legend()
```

```python
import glob
import numpy as np
import matplotlib.pyplot as plt

filenames = sorted(glob.glob('inflammation*.csv'))
plt.figure(figsize=(10.0, 3.0))
plt.ylabel('Difference in average')

for i in range(len(filenames)):
    data0 = np.loadtxt(fname=filenames[i], delimiter=',')
    for j in range(len(filenames)):
        if j > i:
            data1 = np.loadtxt(fname=filenames[j], delimiter=',')
            plt.plot(np.mean(data0, axis=0) - np.mean(data1, axis=0), label=f'{i} and {j}')
            plt.legend()
```
